```python
import sympy as sp
from sympy import symbols, integrate, expand
```

```python
xbars = symbols('\\bar{x}_1 \\bar{x}_2 \\bar{x}_3', real=True)
xbar = sp.Matrix(xbars)
xbar
```

```python
q = symbols('q_1 q_2 q_3 q_4 q_5 q_6 q_7 q_8 q_9 q_10 q_11 q_12', real=True)
q = sp.Matrix(q)
# q[6:12, 0] = sp.zeros(6,1)
q
```

```python
p = sp.Matrix.zeros(3, 1)
A1 = sp.Matrix.zeros(3, 3)
A2 = sp.Matrix.zeros(3, 3)

p = q[0:3, 0]
A1[0, 0:3] = (q[3:6, 0]).T
A1[1, 0:3] = (q[6:9, 0]).T
A1[2, 0:3] = (q[9:12, 0]).T
A2[0:3, 0] = (q[3:6, 0])
A2[0:3, 1] = (q[6:9, 0])
A2[0:3, 2] = (q[9:12, 0])

J1 = dx1dq = (A1 * xbar + p).jacobian(q)
J1
```

```python
J2 = dx2dq = (A2 * xbar + p).jacobian(q)
J2
```

```python
x = symbols('x_1 x_2 x_3', real=True)
x = sp.Matrix(x)

J1.T * x
```

```python
m = symbols('m', real=True)
M1 = m * J1.T * J1
M1
```

```python
M2 = m * J2.T * J2
M2
```

```python
q_tilde = symbols('\\tilde{q}_1 \\tilde{q}_2 \\tilde{q}_3 \\tilde{q}_4 \\tilde{q}_5 \\tilde{q}_6 \\tilde{q}_7 \\tilde{q}_8 \\tilde{q}_9  \\tilde{q}_{10} \\tilde{q}_{11} \\tilde{q}_{12}', real=True)
q_tilde = sp.Matrix(q_tilde)
# q_tilde[6:12, 0] = sp.zeros(6,1)
q_tilde
```

```python
M1 * (q-q_tilde)
```

```python
M2 * (q-q_tilde)
```

```python
H1 = symbols('H11 H21 H31 H12 H22 H32 H13 H23 H33', real=True)
H1 = sp.Matrix(H1)
H = sp.Matrix.zeros(3,3)
H[0:3,0] = H1[0:3]
H[0:3,1] = H1[3:6]
H[0:3,2] = H1[6:9]
H
```

```python
J1.T * H * J1
```

```python
J2.T * H * J2
```

```python
k1 = symbols('k_1', real=True)
k2 = symbols('k_2', real=True)
n = symbols('n_1 n_2 n_3', real=True)
n = sp.Matrix(n)
c1 = symbols('c_1', real=True)
c2 = symbols('c_2', real=True)
w = sp.Matrix.zeros(4,1)
w[0] = c1
w[1] = 1 - c1
w[2] = c2
w[3] = c2 - 1
dtdx = sp.Matrix.zeros(3,12)
dtdx[0:3,0:3] = w[0] * sp.eye(3)
dtdx[0:3,3:6] = w[1] * sp.eye(3)
dtdx[0:3,6:9] = w[2] * sp.eye(3)
dtdx[0:3,9:12] = w[3] * sp.eye(3)
dtdx
```

```python
g = (k1 * n.T * dtdx).T
g
```

```python
H = k2 * ((n.T * dtdx).T * (n.T * dtdx) ).T
H
```

```python
b = sp.Matrix.zeros(12,1)
for ii in range(4):
    gi = g[3*ii:3*(ii+1), 0]
    b = b + J2.T * gi
b
```

```python
A = sp.Matrix.zeros(12,12)
ii = 0; jj = 0
# Hij = J.T * H[3*ii:3*(ii+1), 3*jj:3*(jj+1)]
# Hij
for ii in range(4):
    for jj in range(4):
        Hij = H[3*ii:3*(ii+1), 3*jj:3*(jj+1)]
        A = A + J2.T * Hij * J2
A
```
