```python
import sympy as sp
from sympy import symbols, integrate, expand

x = symbols('x_{1:4}', real=True)
x_tilde = symbols('\\tilde{x}_{1:4}', real=True)
x = sp.Matrix(x)
x_tilde = sp.Matrix(x_tilde)
x - x_tilde
```

```python
mass, dt = symbols('m \\Delta{t}', real=True)
E_inertia = 1/2 * mass / (dt * dt) * (x-x_tilde).T * (x-x_tilde)
E_inertia
```

```python
grad = sp.Matrix([E_inertia]).jacobian(x).T
grad
```

```python
hess = grad.jacobian(x)
hess
```
